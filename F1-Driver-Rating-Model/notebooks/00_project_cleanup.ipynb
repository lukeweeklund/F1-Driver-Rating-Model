```python
from pathlib import Path
import shutil

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

DATA = PROJECT_ROOT / "data"

files_to_delete = [
    DATA / "raw" / "race_results.csv",
    DATA / "raw" / "qualifying_results.csv",
    DATA / "raw" / "race_schedule.csv",
    DATA / "raw" / "circuit_lookup.csv",

    DATA / "intermediate" / "race_results.csv",
    DATA / "intermediate" / "qualifying_results.csv",
    DATA / "intermediate" / "race_schedule.csv",
]

folders_to_delete = [
    DATA / "raw" / "api_cache",
]

print("Deleting files...\n")

for file in files_to_delete:
    if file.exists():
        file.unlink()
        print(f"Deleted: {file.relative_to(PROJECT_ROOT)}")
    else:
        print(f"Not found: {file.relative_to(PROJECT_ROOT)}")

print("\nDeleting folders...\n")

for folder in folders_to_delete:
    if folder.exists():
        shutil.rmtree(folder)
        print(f"Deleted: {folder.relative_to(PROJECT_ROOT)}")
    else:
        print(f"Not found: {folder.relative_to(PROJECT_ROOT)}")

print("\nCleanup complete.")
```
